```python
dfCIC = dfCI.fillna(0)
dfPercent = dfCIC.convert_objects(convert_numeric=True)
dfPercent['Country'] = dfClean.iloc[:,5]
dfCal = dfPercent.groupby('Country').sum()
dfCal.to_csv('item.csv', sep=',')
```

```python
dfCal = dfCal[(dfCal.T != 0).any()]
dfCal.columns.name = 'Factor'
dfCal = dfCal.cumsum()/dfCal.sum()
countries = list(dfCal.index)
factors = list(dfCal.columns)
dfN = pd.DataFrame(dfCal.stack(), columns=['count']).reset_index()
dfN['count'].min()
dfN.head()
```

```python
from math import pi
from bokeh.io import show
from bokeh.models import (
    ColumnDataSource,
    HoverTool,
    LinearColorMapper,
    BasicTicker,
    PrintfTickFormatter,
    ColorBar,
)
from bokeh.plotting import figure, show, output_file

countries = list(dfCal.index)
factors = list(dfCal.columns)

df = pd.DataFrame(dfCal.stack(), columns=['count']).reset_index()

colors = ["#023fa5", "#7d87b9", "#bec1d4", "#d6bcc0", "#bb7784", "#8e063b", "#4a6fe3", "#8595e1", "#b5bbe3", "#e6afb9", "#e07b91", "#d33f6a", "#11c638", "#8dd593", "#c6dec7", "#ead3c6", "#f0b98d", "#ef9708", "#0fcfc0", "#9cded6", "#d5eae7", "#f3e1eb", "#f6c4e1", "#f79cd4"]
mapper = LinearColorMapper(palette=colors, low=df['count'].min(), high=df['count'].max())

source = ColumnDataSource(df)

TOOLS = "hover,save,pan,box_zoom,reset,wheel_zoom"

p = figure(title="Importance of factors that will influence purchase in each country".format(countries[0], countries[-1]),
           x_range=countries, y_range=list(reversed(factors)),
           x_axis_location="above", plot_width=1900, plot_height=400,
           tools=TOOLS, toolbar_location='below')

p.grid.grid_line_color = None
p.axis.axis_line_color = None
p.axis.major_tick_line_color = None
p.axis.major_label_text_font_size = "5pt"
p.axis.major_label_standoff = 0
p.xaxis.major_label_orientation = pi / 3

p.rect(x="Country", y="Factor", width=1, height=1,
       source=source,
       fill_color={'field': 'count', 'transform': mapper},
       line_color=None)

color_bar = ColorBar(color_mapper=mapper, major_label_text_font_size="5pt",
                     ticker=BasicTicker(desired_num_ticks=len(colors)),
                     formatter=PrintfTickFormatter(format="%d%%"),
                     label_standoff=6, border_line_color=None, location=(0, 0))
p.add_layout(color_bar, 'right')

p.select_one(HoverTool).tooltips = [
     ('Country', '@Country'),
     ('count', '@count'),
]

output_file("factors_influence.html")
show(p) 
```
